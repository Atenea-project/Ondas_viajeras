## I) Teoría.

Las ondas son perturbaciones en el espacio que desplazan energía. Estas ondas pueden ser transversales, cuando el cambio en la cantidad física correspondiente ocurre en la dirección del ángulo recto de la dirección de desplazamiento de la onda, o longitudinales, cuando el cambio ocurre a través de la dirección de viaje de la onda.

### 1) Ondas viajeras.

A partir de la función gaussiana

$$y=Ae^{-\frac{x^2}{a^2}}$$

se puede describir una onda que presenta un perfil gaussiano que se desplaza como función del tiempo a través de

\begin{equation}
    y=Ae^{-\frac{(x-vt)^2}{a^2}}
\end{equation}

Cabe notar que una onda mantiene su forma conforme se desplaza.

La forma general para escribir cualquier onda en movimiento es

\begin{equation}
    y=f(x-vt)+g(x+vt)
\end{equation}

y se puede considerar como la superposición de dos ondas cada una a una velocidad $v$ y viajando en direcciones opuestas. La deducción de la ecuación previa se describe a mayor detalle en la sección 5.4 de (King, 2009).


#### 1.1) Ondas viajeras periodicas.

Formas de ondas muy complejas pueden ser descritas mediante la combinación de ondas sinuisodales.
Las ondas sinuisodales consisten en patrones repetidos y la distancia entre puntos similares del patron repetido es la longitud de onda $(\lambda)$.

\begin{equation}
    y_{(x,t)}=A\sin{\frac{2\pi}{\lambda}}(x-vt)
\end{equation}

La frecuencia $(\nu)$ es el número de veces, por unidad de tiempo, que la cresta de una onda pasa por un punto fijo.

$$\nu\lambda=v$$

$$\nu=\frac{1}{T}$$

Podemos expresar la ecuación de la onda sinuisodal de la siguiente forma

$$y(x_0,t)=-A\sin\frac{2\pi}{\lambda}(vt-x_0)$$

usando la relación $\sin(\alpha-\beta)=-\sin(\beta-\alpha)$ obtenemos

\begin{equation}
    \omega=\frac{2\pi v}{\lambda}=2\pi\nu
\end{equation}

Definimos el número de onda $k$

\begin{equation}
    k=\frac{2\pi}{\lambda}
\end{equation}

Con lo que nos queda

\begin{equation}
    y(x,t)=Asen(kx-\omega t)
\end{equation}

\begin{equation}
    v=\frac{\omega}{k}
\end{equation}

Puede ser ventajoso expresar esta función en terminos complejos

$$y(x,t)=Ae^{\frac{2\pi}{\lambda}i(x-vt)}$$

$$=Ae^{i\left(\frac{2\pi}{\lambda}x-\frac{2\pi}{\lambda}vt\right)}$$

$$=Ae^{i(kx-\omega t)}$$

### 2) Energía de una onda.

Consideramos el caso para ondas transversales en una cuerda tensa. Imaginemos pequeños fragmentos de ancho $\delta x$ y masa $\mu\delta x$, donde $\mu$ es la masa por unidad de longitud. Conforme la onda se mueve por la cuerda estos segmentos oscilan en dirección transversal y su energía cinética estara dada por

\begin{equation}
    K=\frac{1}{2}\mu\delta x\left(\frac{\partial y}{\partial t}\right)^2
\end{equation}

La longitud extendida se aproxima de la siguiente manera

$$\delta s=\frac{\delta x}{\cos\theta}=\frac{\delta x}{\sqrt{1-\sin^2\theta}}$$

Cuando $\theta$ es pequeño

$$\delta s\approx\frac{\delta x}{\theta}\approx\delta x\left(1+\frac{1}{2}\theta^2\right)$$

tal que 

$$\theta=\frac{\partial y}{\partial x}$$

Con lo que nos queda

$$\delta s\approx\delta x\left(1+\frac{1}{2}\left(\frac{\partial y}{\partial x}\right)^2\right)$$

Una buena aproximación a la energía potencial seria 

\begin{equation}
    U=T(\delta s-\delta x)=\frac{1}{2}T\delta x \frac{\partial^2y}{\partial x^2}
\end{equation}

Con lo que podemos escribir una ecuación que describa la energía para una sección $a\leq x\leq b$ para una cuerda, al tiempo t

\begin{equation}
    E=\frac{1}{2}\int_a^bdx\left[\mu\frac{\partial^2y}{\partial t^2}+T\frac{\partial^2}{\partial x^2}\right]=\frac{1}{2}\mu\int_a^bdx\left[\frac{\partial^2y}{\partial t^2}+v^2\frac{\partial^2}{\partial x^2}\right]
\end{equation}

Definimos a la velocidad de propagación de la onda como

\begin{equation}
    v=\sqrt{\frac{T}{\mu}}
\end{equation}

En el caso de una onda sinuisodal tenemos 

$$y=A\sin(kx-\omega t)$$

$$K=\frac{1}{2}\mu\delta x\frac{\partial^2y}{\partial t^2}=\frac{1}{2}\mu\delta x\omega^2A^2\cos^2(kx-\omega t)$$

La energía cinética total contenida en una longitud de onda $\lambda$ es

$$K_{total}=\frac{1}{2}\mu\omega^2A^2\int_0^\lambda\cos^2(kx-\omega t)dx=\frac{1}{2}\mu\omega^2A^2\left(\frac{\lambda}{2}\right)$$

\begin{equation}
    K_{total}=\frac{1}{4}\mu\omega^2A^2\lambda
\end{equation}

En el caso de la energía potencial tenemos 

$$U=\frac{1}{2}v^2\mu dx\left(\frac{\partial y}{\partial x}\right)^2=\frac{1}{2}v^2\mu\delta xk^2A^2\cos^2(kx-\omega t)=\frac{1}{2}\mu\delta x\omega^2A^2\cos^2(kx-\omega t)$$

Recordando que

$$v=\frac{k}{x}$$

Con lo que nos quedan las siguientes ecuaciones que aplican para todos los valores de $t$

\begin{equation}
    U_{total}=\frac{1}{4}\mu\omega^2A^2\lambda
\end{equation}

\begin{equation}
    E_{total}=\frac{1}{2}\mu\omega^2A^2\lambda
\end{equation}

#### 2.1) El transporte de energía por una onda.

Partiendo de la ecuación para la energía de la onda sabemos que la energía esta asociada con la derivada de desplazamiento con respecto al tiempo y la posción.
La energía llevada por una onda a través de cualquier linea en angulo recto a la dirección de propagación por unidad de tiempo es la potencia.

\begin{equation}
    P=\frac{1}{2}\mu\omega^2A^2v
\end{equation}

## II) Código, ondas viajeras.
Crearemos una función con la que podemos graficar la variación de la ecuación de la onda con respecto a sus parametros y el tiempo.

Para hecer más visual a la onda construiremos un pequeño video que refleja el desplazamiento de la onda.

Primero importamos las librerías adecuadas 

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import animation
from IPython.display import HTML
from ipywidgets import interact
```

Crearemos una serie de funciones para describir el comportamiento de la onda a través del tiempo y visualizar este comportamiento en una animación.

Lo primero que haremos es crear el fondo de la animación, que sera una gráfica vacia de las dimensiones de interes. 

Las funciones usadas son:

  - plt.subplots(), que gráfica varios elementos en una misma gráfica.
  - 'z'.set_xlim(('x','y')), fija los límites del eje X para la gráfica 'z' de 'x' a 'y'. 
  - 'z'.set_ylim(('x','y')), fija los límites del eje Y para la gráfica 'z' de 'x' a 'y'. 
  - 'z'.plt(['x'],['y'],lw='a'), que gráfica los valores de ['y'] con respecto a ['x'] usando un grueso de linea (linewidth) de 'a'.

Si quieres entender mejor estos parametros puedes consultar la documentación oficial: [matplotlib.pyplot.plot](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot)

```python
figura, ejes = plt.subplots()

ejes.set_xlim(( 0, 500))
ejes.set_ylim((-2, 20))

linea, = ejes.plot([], [], lw=3)
```

Ahora creamos una función 'init()' que nos regrese una gráfica base para la animación.

```python
def init():
    linea.set_data([], [])
    return (linea,)
```

Usaremos la siguiente función para describir la onda

\begin{equation}
    y=Ae^{-\frac{(x-vt)^2}{a^2}}
\end{equation}

Que describe un comportamiento gausiano para la onda. 

En esta función los parámetros que les pasaremos son

    - 'A': Amplitud de la onda.
    - 'v': Velocidad de la onda.
    - 'a': Parametro asociado a la anchura de la onda.

también tenemos que generar un espacio lineal de los tiempos 'x' para evaluar a las funciones.

**Para esta función se usara el parametro 'i' que actuara como el tiempo y esta asociado a la función de la animación.**

Las funciones que pueden ser útiles son:

    - np.linspace(x,y,z), que genera un espacio lineal desde 'x' hasta 'y' con 'z' elementos.
    - np.exp(x), que calcula la potencia 'x' de e.
    - 'a'.set_data('x','y'), fija los datos 'x' e 'y' a la gráfica 'a'.

Despues construiremos la función 'animac_onda' que tomara los parametros 'A' asociado a la amplitud de la onda, 'v' asociado a la velocidad de la onda y 'a' relacionado a la anchura de la onda. Esta función generara una animación de la función 'onda'

Las funciones que pueden ser útiles son:

    - animation.FuncAnimation('w', 'x', init_func='y',
                               frames='z', interval='a', fargs='b',
                               blit='c'), 
      que genera la animación de la función 'x' cuyos parametros son una tupla 'b', usando como base una figura 
      'w' y una gráfica 'y' que requiere un iterable correspondiente al número de cuadros 'z' en la animación 
      que tendran un retraso entre cuadro de 'a' milisegundos. Por último el parametro 'blit' esta asociado a un 
      valor booleano (True, False) que en caso de estar activado optimiza el dibujo de la animación.

Al final la función 'animac_onda' regresara una animación en forma de video de HTML5. 

Si quieres entender mejor estos parametros puedes consultar la documentación oficial: [matplotlib.animation.FuncAnimation](https://matplotlib.org/3.3.2/api/_as_gen/matplotlib.animation.FuncAnimation.html)

```python
def onda(i,A,v,a):
    x = np.linspace(0, 500, 1000)
    y = A*np.exp(-((x-v*i)**2)/a**2)
    linea.set_data(x, y)
    return (linea,)
```

```python
def animac_onda(A,v,a):
    ond_anim = animation.FuncAnimation(figura, onda, init_func=init,
                               frames=100, interval=40, fargs=(A,v,a),
                               blit=True)
    return(HTML(ond_anim.to_html5_video()))
```

Generamos una gráfica interactiva con la función 'animac_onda'

- interact(w,x=(a,b),...,z=(s,d)), genera una gráfica interactiva de la función 'w' donde los parámetros de la función de 'x' varían de 'a' a 'b', hasta 'z' varían de 's' a 'd'.

en este caso te recomendamos que 'A' y 'v' tengan valores de 0 a 10 y que 'a' vaya de 1 a 20.

**¡NOTA!** Dado que en este caso estamos generando una animación debes de esperar unos segundos para que la nueva animación carge cada vez que cambies un parametro.


```python
interact(animac_onda,A=(0,10),v=(0,10),a=(1,20))
```

### 3) Ondas y discontinuidades..


Cuando una onda se encuentra con una discontinuidad en la frontera entre 2 medios, por lo general una fracción de la onda se refleja, generando 3 ondas.

    1.Onda incidente.
    2.Onda transmitida.
    3.Onda reflejada.

La velocidad de onda esta dada por $v=\frac{\omega}{k}=\sqrt{T/\mu}$, donde $T$ es la tensión de la cuerda y $\mu$ es la masa por unidad de longitud, como a cada lado de la discontinuidad hay diferentes valores de masa por unidad de longitud $\mu$, tendremos un cambio de velocidades. Cabe destacar que a lo largo de la discontinuidad la frecuencia $\omega$ es la misma para ambos fragmentos de onda.

Para modelar a las 3 ondas se plantean dos cuerdas unidas, cada cuerda tiene su valor de masa por unidad de longitud $\mu$, lo que genera una discontinuidad.
Se contemplan las siguientes condiciones:

**A)** La frecuencia $\omega$ es la misma para ambas cuerdas, la longitud de onda sera diferente para ambas cuerdas, dada
   
$$\lambda=\frac{2\pi v}{\omega}$$

**B)** $\frac{\partial y}{\partial x}$ deben de ser iguales en $x=0$ para ambas cuerdas en cualquier momento, dado que la fuerza transversal es igual a $T\frac{\partial y}{\partial x}$, como $T$ es constante, las pendientes $\frac{\partial y}{\partial x}$ de las cuerdas en ambos lados de la union deben de ser iguales en $x=0$, que es la posición de union, para cualquier tiempo $t$.

Definimos

$$y_I=A_1\cos(\omega t -k_1x)\quad\textbf{Onda incidente}$$

$$y_T=A_2\cos(\omega t -k_2x) \quad\textbf{Onda transmitida}$$

$$y_R=B_1\cos(\omega t +k_1x) \quad\textbf{Onda reflejada}$$

$$y_1=y_I+y_R , y_2=y_T$$


En $x=0$ se cumple que $y_1=y_2$, por lo que podemos escribir.


$$A_1\cos(\omega t -k_1x)+B_1\cos(\omega t +k_1x)=A_2\cos(\omega t -k_2x)$$

Dado que en $t=0$ se cumple que $x=0$

\begin{equation}
    A_1+B_1=A_2
\end{equation}

Recordando la segunda condición, obtenemos $\frac{\partial y_1}{\partial x}=\frac{\partial y_2}{\partial x}$ en $x=0$ para cualquier $t$.

$$k_1A_1\sin(\omega t-k_1x)-k_1B_1\sin(\omega t+k_1x)=k_2A_2\sin(\omega t-k_2x)$$

Si $x=0, t=\frac{\pi}{2\omega}$ y como $\sin\left(\frac{\pi}{2}\right)=1$

\begin{equation}
    k_1A_1-k_1B_1=k_2A_2
\end{equation}

Usando las relaciones entre las constantes, podemos definir al coeficiente de transmisión de la amplitud ($T_{12}$), que adquiere valores entre 0 y 2.

$$\frac{A_2}{A_1}=\frac{2k_1}{k_1+k_2}=T_{12}$$

Tambien podemos definir al coeficiente de reflexíon de la amplitud ($R_{12}$), que adquiere valores entre -1 y 1.

$$\frac{B_1}{A_1}=\frac{k_1-k_2}{k_1+k_2}=R_{12}$$

Recordando que $v=\frac{\omega}{k}=\sqrt{\frac{T}{\mu}}$

$$T_{12}=1+R_{12}$$

\begin{align}
    T_{12}=\frac{2\sqrt{\mu_1}}{\sqrt{\mu_1}+\sqrt{\mu_2}}\\
    R_{12}=\frac{\sqrt{\mu_1}-\sqrt{\mu_2}}{\sqrt{\mu_1}+\sqrt{\mu_2}}
\end{align}

Si $\mu_2\rightarrow\infty \Rightarrow A_2=0, B_1=-A_1$ por lo que si una onda se encuentra con una pared sólida, toda la onda sera reflejada.

#### 3.1) Indices de refracción.

Definimos al indice de refracción en base a la vecidad de la luz en el vacio ($c$) y la velocidad de la luz en el medio ($v$) del que queremos obtener su indice de refracción.

\begin{equation}
    \eta=\frac{c}{v}
\end{equation}

## II.I) Código discontinuidad en una onda.

Crearemos una función 'disc' que en base a la información de una onda y las masas por unidad de longitud de ambas secciones de la discontinuidad genere una animación en la que podamos ver este fenomeno físico.




```python
figura1, ejes1 = plt.subplots()

ejes1.set_xlim(( -200, 200))
ejes1.set_ylim((-25, 25))

linea1, = ejes1.plot([], [], lw=1)
linea2, = ejes1.plot([], [], lw=1)
linea3, = ejes1.plot([], [], lw=1)
linea4, = ejes1.plot([], [], lw=1)
```

```python
def init1():
    linea1.set_data([], [])
    linea2.set_data([], [])
    linea3.set_data([], [])
    linea4.set_data([], [])
    return (linea1,)
```

Las funciones matemáticas útiles son

$$y_I=A_1\cos(\omega t -k_1x)\quad\textbf{Onda incidente}$$

$$y_T=A_2\cos(\omega t -k_2x) \quad\textbf{Onda transmitida}$$

$$y_R=B_1\cos(\omega t +k_1x) \quad\textbf{Onda reflejada}$$

El coeficiente de transmisión de la amplitud ($T_{12}$), que adquiere valores entre 0 y 2

$$\frac{A_2}{A_1}=\frac{2k_1}{k_1+k_2}=T_{12}$$

$$T_{12}=\frac{2\sqrt{\mu_1}}{\sqrt{\mu_1}+\sqrt{\mu_2}}$$

$$A_2=\frac{A_1\cdot2\sqrt{\mu_1}}{\sqrt{\mu_1}+\sqrt{\mu_2}}$$

Las otras relaciones que nos serviran, son

$$B_1=A_2-A_1$$

y para el valor de $k_2$

$$k_2=\frac{k1\cdot(A1-B1)}{A2}$$

a la función 'disc' le pasaremos los valores $A1$ amplitud dela onda incidente, $w$ frecuencia de las ondas (recordar que las tres ondas tendran la misma frecuencia), $k1$ número de onda de la onda incidente ($k=\frac{2\pi}{\lambda}$), $m1$ masa por unidad de longitud de la primer onda ($\mu_1$) y $m2$ masa por unidad de longitud de la segunda onda ($\mu_2$).

Las funciones que te pueden ser de útilidad son:

  - np.sqrt('x'), que te da el valor de la riz cuadrada de 'x'.
  - np.cos('x'), que te da el valor del coseno de 'x', esta función toma a 'x' en radianes.

```python
def disc(i,A1,w,k1,m1,m2):
    leyenda='A1: {}, w: {}, k1: {}, m1: {}, m2: {}'.format(A1,
                                                         '{0:.2g}'.format(w),
                                                         '{0:.2g}'.format(k1),
                                                         m1,m2)
    x1 = np.linspace(-300, 0, 1000)
    x2=np.linspace(0, 300, 1000)
    A2=A1*(2*np.sqrt(m1))/(np.sqrt(m1)+np.sqrt(m2))
    B1=A2-A1
    k2=k1*(A1-B1)/A2
    k2=k1/2
    y1 = A1*np.cos(w*i-k1*x1)
    linea1.set_data(x1, y1)
    linea1.set_label('Onda incidente')
    y2= A2*np.cos(w*i-k2*x2)
    linea2.set_data(x2, y2)
    linea2.set_label('Onda Transmitida')
    y3=B1*np.cos(w*i+k1*x1)
    linea3.set_data(x1,y3)
    linea3.set_label('Onda Reflejada')
    linea4.set_label(leyenda)
    ejes1.legend()
    return (linea1,)
```

Por último programaremos la animación de estas ondas. 

Solo hace falta que ejecutes el código de las siguientes celdas.

```python
def animac_disc(A1,w,k1,m1,m2):
    disc_anim = animation.FuncAnimation(figura1, disc, init_func=init1,
                               frames=150, interval=80, fargs=(A1,w,k1,m1,m2),
                               blit=True)
    return(HTML(disc_anim.to_html5_video()))
```

```python
interact(animac_disc,A1=(1,10),w=(0.1,0.5),k1=(0.1,0.5),m1=(1,10),m2=(1,10),)
```
